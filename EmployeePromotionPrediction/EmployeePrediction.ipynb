### FEATURE ENGINEERING

```python
data['sum_metric'] = data['previous_year_rating'] + data['KPIs_met >80%'] + data['awards_won?']
data['total_score'] = data['avg_training_score'] + data['no_of_trainings']
```

```python
data.drop(['region' , 'recruitment_channel'] , axis = 1 , inplace = True)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data['department'] = le.fit_transform(data['department'])
```

```python
data['gender'] = le.fit_transform(data['gender'])
```

```python
data['education'] = data['education'].replace(('Master\'s & above' , 'Bachelor\'s' , 'Below Secondary'),
                                             (3,2,1))
```

```python
data.head()
```

```python
df = data.drop(['previous_year_rating' , 'KPIs_met >80%' , 'awards_won?' , 'avg_training_score' , 'no_of_trainings'] , axis = 1)
```

```python
df.head()
```

```python
X = df.drop('is_promoted' , axis = 1)
Y = df['is_promoted']

X.shape , Y.shape
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state = 2)
X_res , Y_res = sm.fit_resample(X , Y)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_res = sc.fit_transform(X_res)
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , Y_train , Y_test = train_test_split(X_res,Y_res ,
                                                       test_size = 0.2 , random_state = 2)
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score , confusion_matrix, roc_curve ,roc_auc_score,precision_score,recall_score , auc
```

```python
model = DecisionTreeClassifier()
model.fit(X_train , Y_train)
```

```python
model.score(X_test , Y_test)
```

```python
pred = model.predict(X_test)
```

```python
accuracy_score(Y_test , pred)
```

```python
print(classification_report(Y_test , pred))
```

```python
sns.heatmap(confusion_matrix(Y_test , pred) , annot = True)
```

```python
confusion_matrix(Y_test , pred)
```

```python
df['is_promoted'].value_counts()
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(X_train , Y_train)

model.score(X_test , Y_test)
```

```python
pred = model.predict(X_test)
```

```python
print(classification_report(Y_test , pred))
```

```python
sns.heatmap(confusion_matrix(Y_test , pred) , annot = True , fmt ='.4g')
```

```python
r = roc_curve(Y_test , pred)

plt.plot(r[1] , r[0])
```

```python
pred_probs = model.predict_proba(X_test)
```

```python
pred_probs.shape
```

```python
prob_1 = pred_probs[: , 1]   # prob of class 1
prob_0 = pred_probs[: , 0]   # prob of class 0
```

```python
FPR , TPR , _ = roc_curve(Y_test , prob_1) # we are doing it only for "1"

plt.plot(FPR , TPR , marker = '*')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```

```python
FPR , TPR , _ = roc_curve(Y_test , prob_0) # we are doing it only for "1"

plt.plot(FPR , TPR , marker = '*')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```

```python
area = auc(FPR , TPR)
```

```python
Auc = roc_auc_score(Y_test , prob_1)
print("Area Under Curve: " , Auc)
```

```python
plt.plot(FPR , TPR , 'b' , label = 'AUC = %0.2f' % area)
plt.legend()
```
