```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cols = ['id_user' , 'item_id' , 'rate' , 'timestamp']
```

```python
data = pd.read_csv(r'u.data' , sep = '\t' , names = cols)
```

```python
data.head()
```

```python
movies = pd.read_csv('Movie_names.txt' , sep = ',')
movies.head()
```

```python
data = pd.merge(data , movies , on = 'item_id')
data.head()
```

```python
data.groupby('title')['rate'].mean().sort_values(ascending = False).head()
```

```python
data.groupby('title')['rate'].count().sort_values(ascending = False).head()
```

```python
ratings = pd.DataFrame(data.groupby('title')['rate'].mean())
```

```python
ratings.head()
```

```python
ratings['no_of_ratings'] = pd.DataFrame(data.groupby('title')['rate'].count())
```

```python
ratings.head()
```

```python
ratings['no_of_ratings'].hist(bins = 70)
```

```python
ratings['rate'].hist(bins = 70)
```

```python
sns.jointplot(data = ratings , x = 'rate' , y = 'no_of_ratings' , alpha = 0.5)
```

```python
moviemat = data.pivot_table(index = 'id_user' , columns = 'title' , values = 'rate')
moviemat.head()
```

```python
ratings.sort_values('no_of_ratings' , ascending = False).head(10)
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
liarliar_user_ratings = moviemat['Liar Liar (1997)']
```

```python
starwars_user_ratings.head()
```

```python
moviemat.corrwith(starwars_user_ratings)
```

```python
similar_to_starwar = moviemat.corrwith(starwars_user_ratings) 
```

```python
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
corr_starwar = pd.DataFrame(similar_to_starwar , columns = ['Correlation'])
corr_starwar.dropna(inplace = True)
corr_starwar.head()
```

```python
corr_starwar.sort_values(ascending = False , by = 'Correlation').head(10)
```

```python
corr_starwar = corr_starwar.join(ratings['no_of_ratings'])
corr_starwar.head()
```

```python
corr_starwar[corr_starwar['no_of_ratings'] > 100].sort_values(ascending = False , by = 'Correlation').head(10)
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar , columns = ['Correlation'])
corr_liarliar.dropna(inplace = True)
corr_liarliar.head()
```

```python
corr_liarliar = corr_liarliar.join(ratings['no_of_ratings'])
corr_liarliar.head()
```

```python
corr_liarliar[corr_liarliar['no_of_ratings'] > 100].sort_values(ascending = False , by = 'Correlation').head(10)
```
